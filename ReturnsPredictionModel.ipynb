```python
import pandas as pd


train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
sample = pd.read_csv("SampleSubmission.csv")


train.head()
```

```python

train = train.copy()
test = test.copy()

train['Item_Weight'] = train['Item_Weight'].fillna(train['Item_Weight'].mean())
test['Item_Weight'] = test['Item_Weight'].fillna(train['Item_Weight'].mean())

train['Store_Size'] = train['Store_Size'].fillna(train['Store_Size'].mode()[0])
test['Store_Size'] = test['Store_Size'].fillna(train['Store_Size'].mode()[0])


from sklearn.preprocessing import LabelEncoder

cols_to_encode = ['Item_Sugar_Content', 'Item_Type', 'Store_Size', 'Store_Location_Type', 'Store_Type']
le = LabelEncoder()

for col in cols_to_encode:
    train[col] = le.fit_transform(train[col])
    test[col] = le.transform(test[col])
```

```python
from sklearn.ensemble import RandomForestRegressor

# Prepare input features
X = train.drop(['Item_Store_Returns', 'Item_ID', 'Store_ID', 'Item_Store_ID'], axis=1)
y = train['Item_Store_Returns']

# Prepare test data with same features
X_test = test.drop(['Item_ID', 'Store_ID', 'Item_Store_ID'], axis=1)


model = RandomForestRegressor(n_estimators=100, random_state=42)


model.fit(X, y)
```

```python
# Predict the target for test data
predictions = model.predict(X_test)

# Copy the sample submission file
submission = sample.copy()

# Add predictions to the 'Item_Store_Returns' column
submission['Item_Store_Returns'] = predictions

# Save the submission to a CSV file
submission.to_csv('submission.csv', index=False)

```
